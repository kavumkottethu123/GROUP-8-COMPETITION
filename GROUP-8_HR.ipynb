##### Insights:
* education is a categorical feature and contain null values.It also signifies whether the person has obtained a particular level of education or not, we cannot impute the null values as we can't assign an education level to a person on our own. The person may or may not have achieved the assigned level and this will result in an inaccurate analysis
* previous_year_rating is a numerical feature and contain null values.The NULL value in this column indicates that the employee has been a part of the company for LESS THAN 1 YEAR and does not have a previous year score. Hence we can retain these null values with 0

##### Splitting the dataset

```python
x=hr_data.drop(['region','department','awards_won?'],axis=1)
y=hr_data['awards_won?']
```

```python
x.head()
```

```python
#Onehot Encoding
x['education'].value_counts()
```

```python
x['gender'].value_counts()
```

```python
edu_en=pd.get_dummies(x['education'])
gender_en=pd.get_dummies(x['gender'])
recruitment_en=pd.get_dummies(x['recruitment_channel'])
```

```python
x=pd.concat([x,edu_en,gender_en,recruitment_en],axis=1)
```

```python
x=x.drop(['education','gender','recruitment_channel'],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.3)
```

```python
x_train.shape
```

#### Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred=logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
print("Accuracy is:",accuracy_score(y_test,y_pred))
print("Precision is:",precision_score(y_test,y_pred))
print("Recall is:",recall_score(y_test,y_pred))
print("F1 score is:",f1_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
acc_values=[]
neighbors=np.arange(3,15)
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski')
    classifier.fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    acc_values.append(acc)
```

```python
plt.plot(neighbors, acc_values,'0.')
plt.xlabel('k value')
plt.ylabel('accuracy')
```

### Modelling

#### Splitting the dataset

```python
x=hr_data.drop(['region','department','awards_won?'],axis=1)
y=pd.DataFrame(hr_data['awards_won?'])
```

```python
edu_en=pd.get_dummies(x['education'])
gender_en=pd.get_dummies(x['gender'])
recruitment_en=pd.get_dummies(x['recruitment_channel'])
```

```python
x=pd.concat([x,edu_en,gender_en,recruitment_en],axis=1)
```

```python
x=x.drop(['education','gender','recruitment_channel'],axis=1)
```

```python
x.head()
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.3)
```

#### Random Forest

##### Build the model

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import f1_score,confusion_matrix
f1_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#Feature importance
pd.Series(rf.feature_importances_,index=x.columns).sort_values(ascending=False)*100
```

```python
#Distribution of prediction probabilities
threshold=0.22
y_pred_prob=rf.predict_proba(x_test)[:,1]
y_pred=(y_pred_prob>threshold).astype(int)
```

```python
f1_score(y_test,y_pred)
```
